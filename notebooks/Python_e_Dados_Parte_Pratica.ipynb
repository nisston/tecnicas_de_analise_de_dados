# Limpar os dados

```python
# Verificar valores ausentes
print("\nValores ausentes por coluna:")
print(df.isnull().sum())
```

```python
# Preencher valores nulos (exemplo: substituir por 0 onde fizer sentido)
df.fillna({'mandante_placar': 0, 'visitante_placar': 0}, inplace=True)
```

```python
# Remover duplicatas caso existam
df.drop_duplicates(inplace=True)
```

# Transformar os dados

```python
df
```

```python
df.columns
```

```python
# Criar uma nova coluna de saldo de gols
df['saldo_gols'] = df['mandante_placar'] - df['visitante_placar']
```

```python
# Criar uma coluna categórica indicando se o mandante venceu, empatou ou perdeu
def classificar_resultado(row):
    if row['saldo_gols'] > 0:
        return 'Vitória Mandante'
    elif row['saldo_gols'] < 0:
        return 'Vitória Visitante'
    else:
        return 'Empate'
```

```python
df['resultado'] = df.apply(classificar_resultado, axis=1)
```
